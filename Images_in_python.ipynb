```python
# Image:
input_img = Image.open("Circle_2.jpg")
input_pixels = input_img.load()
width, height = input_img.width, input_img.height

# Create output image:
output_img = Image.new("RGB", input_img.size)
draw = ImageDraw.Draw(output_img)

# Convert to grayscale:
intensity = np.zeros((width, height))
for x in range(width):
    for y in range(height):
        intensity[x, y] = sum(input_pixels[x, y]) / 2

# Compute convolution between intensity and kernels:
for x in range(1, input_img.width - 1):
    for y in range(1, input_img.height - 1):
        magx = intensity[x + 1, y] - intensity[x - 1, y]
        magy = intensity[x, y + 1] - intensity[x, y - 1]
        color = int(sqrt(magx**2 + magy**2))
        draw.point((x, y), (color, color, color))
    
plt.imshow(output_image, cmap='gray')
```
